# 3. How to use color palette

# 3a. global set

```python
# Path of the file to read
flight_filepath = "input/flight_delays.csv"

# Read the file into a variable flight_data
flight_data = pd.read_csv(flight_filepath)
```

## set "Paired"

```python
sns.set_palette("Paired")
```

```python
# Set the width and height of the figure
plt.figure(figsize=(10,6))

# Add title
plt.title("Average Arrival Delay for Spirit Airlines Flights, by Month")

# Bar chart showing average arrival delay for Spirit Airlines flights by month
sns.barplot(x = flight_data.index, y = flight_data['NK'])
#sns.barplot(x = flight_data.index, y=flight_data['AA'])

# Add label for vertical axis
plt.ylabel("Arrival delay (in minutes)")
```

## Reset default

```python
sns.set()
sns.set_style("white") # reset sns 
```

```python
# Set the width and height of the figure
plt.figure(figsize=(10,6))

# Add title
plt.title("Average Arrival Delay for Spirit Airlines Flights, by Month")

# Bar chart showing average arrival delay for Spirit Airlines flights by month
sns.barplot(x = flight_data.index, y = flight_data['NK'])

#sns.barplot(x = flight_data.index, y=flight_data['AA'])

# Add label for vertical axis
plt.ylabel("Arrival delay (in minutes)")
```

# 3b. local set

```python
with sns.color_palette("Set3"): # number of colors cant be execeed this, otherwise default
    
    # Set the width and height of the figure
    plt.figure(figsize=(10,6))

    # Add title
    plt.title("Average Arrival Delay for Spirit Airlines Flights, by Month")

    # Bar chart showing average arrival delay for Spirit Airlines flights by month
    sns.barplot(x = flight_data.index, y = flight_data['NK'])

    #sns.barplot(x = flight_data.index, y=flight_data['AA'])

    # Add label for vertical axis
    plt.ylabel("Arrival delay (in minutes)")
```
